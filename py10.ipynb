# 영상필터링

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('C:/BSW/frame_choice2.jpg')
img.shape

blurred = cv2.GaussianBlur(img, (7,7),0)
edge = cv2.Canny(blurred, 50, 150)


plt.subplot(1,2,1)
plt.imshow(img, cmap="gray")
plt.title("ori")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(edge, cmap="gray")
plt.title("edge")
plt.axis("off")
plt.show
```
